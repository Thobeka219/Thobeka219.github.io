```python
# removing the item_Identifier since these are just the unique values
train = train.drop('Item_Identifier', axis=1)
```

```python
train.head()
```

```python
# separating the independent and dependent variables

# storing all the independent variables as X
X_Train = train.drop('Item_Outlet_Sales', axis=1)

# storing the dependent variable as y
y_Train = train['Item_Outlet_Sales']
```

```python
# shape of independent and dependent variables
X_Train.shape, y_Train.shape
```

```python
# Creating training and validation set

X_train,X_test,y_train,y_test = train_test_split(X_Train,y_Train,random_state=20,test_size=0.3)
```

```python
# shape of training and validation set
(X_train.shape, y_train.shape), (X_test.shape, y_test.shape)
```

```python
print(f'X_train training set {X_train.shape}\nX_test testing set {X_test.shape}\n y_train training set {y_train.shape}\ny_test testing set {y_test.shape}')
```

```python
print(keras.__version__)
```

```python
# number of input neurons
X_train.shape
```

```python
# number of features in the data
X_train.shape[1]
```

```python
# defining input neurons
input_neurons = X_train.shape[1]
```

```python
# define number of output neurons
output_neurons = 1
```

```python
# define hidden layers and neuron in each layer
number_of_hidden_layers = 2
neuron_hidden_layer_1 = 40
neuron_hidden_layer_2 = 20
```

```python
# defining the architecture of the model
model = Sequential()
model.add(Dense(80, input_dim=22, activation= "relu"))
model.add(Dense(units=neuron_hidden_layer_1, activation='relu'))
model.add(Dense(units=neuron_hidden_layer_2, activation='relu'))
#5. Since it is a regression problem, make sure that you use the 'linear' activation function in the output layer.
model.add(Dense(units=output_neurons, activation='linear'))
```

```python
# summary of the model
model.summary()
```

```python
# number of parameters between input and first hidden layer

input_neurons*neuron_hidden_layer_1
```

```python
# number of parameters between input and first hidden layer

# adding the bias for each neuron of first hidden layer

input_neurons*neuron_hidden_layer_1 + 40
```

```python
# number of parameters between first and second hidden layer

neuron_hidden_layer_1*neuron_hidden_layer_2 + 20
```

```python
# number of parameters between second hidden and output layer

neuron_hidden_layer_2*output_neurons + 1
```

```python
# compiling the model
model.compile(loss='mse', optimizer='Adam', metrics=['mse','mae', 'accuracy'])
```

**Question 4:TRAINING THE MODEL**

```python

model_history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50)
```

```python
# getting predictions for the validation set
predictions = np.argmax(model.predict(X_test),axis=1)
```

```python
# calculating the accuracy on validation set
from sklearn import linear_model
regressor = linear_model.LinearRegression()
regressor.fit(X_train,y_train)
prediction = regressor.predict(X_test)
# accuracy_score(y_test, predictions)
```

```python
from numpy import math
from sklearn.metrics import mean_squared_error

math.sqrt(mean_squared_error(y_test,prediction))
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, prediction)
```

```python
# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
plt.plot(model_history.history['mse'])
plt.plot(model_history.history['val_mse'])
plt.title('model mse')
plt.ylabel('mse')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
